# Who Are These Notes For?

This Jupyter Book is intended for students who are beginning to study astronomy, planetary science, astrobiology, or related sciences. It is written for readers who may be new to scientific reasoning, quantitative thinking, or computational tools, but who are willing to ask questions and think carefully.

You do not need to begin as an expert. Curiosity, attention, and steady effort matter more than prior experience.

---

## What This Book Is

This book is a guided set of notes for ASTR120. It is designed to help you understand major ideas in astronomy and astrobiology, especially how scientific claims are connected to evidence.

The book emphasizes three connected habits:

1. **Scientific reasoning** — asking what evidence supports a claim.
2. **Quantitative reasoning** — using basic numbers and comparisons to make ideas more concrete.
3. **Curious exploration** — using figures, examples, optional code, and AI study tools to deepen your understanding.

The goal is not only to learn what scientists currently think. The goal is to understand why scientists think it.

---

## What This Book Is Not

This book is not a list of facts to memorize. It is also not a programming textbook or a math-heavy problem-solving manual.

Instead, it should be used as a working resource. You should read actively, pause at examples, think through checkpoint questions, and return to earlier sections when a later idea depends on them.

Some sections include simple calculations or optional Python examples. These are included to support understanding, not to make the course focused on coding or advanced mathematics.

Science is learned by asking questions, comparing explanations, and testing ideas against evidence. Treat this book as a place to practice that process.

---

## Interacting with the Book

Many sections of this book are written as Jupyter notebooks. This means that some pages may include code cells, figures, calculations, or interactive elements.

You do not need prior Python experience to use these notes. When Python appears, it is there for interested students who want to explore a calculation, graph, or idea more deeply. The most important goal is to understand the scientific idea being shown.

Each chapter can be opened using [Google Colab](https://research.google.com/colaboratory/) through the rocket icon at the top of the page.

You are welcome to run code cells, change values, and see what happens. If a result changes, ask why. If a graph looks different than expected, ask what scientific relationship the graph is showing.

You may also use AI tools, including NotebookLM, to study these notes. These tools can help summarize sections, explain vocabulary, generate practice questions, or check your understanding. They should support your learning, not replace your own thinking.

---

## How to Use This Book Well

A useful way to study is to move through each section in stages:

1. Read for the main idea first.
2. Ask what scientific claim is being made.
3. Look for the evidence that supports the claim.
4. Pause at examples, figures, and checkpoint questions.
5. Use optional calculations, code, or AI tools to explore ideas more deeply.

You will get more out of this book by working with it than by simply reading it.
