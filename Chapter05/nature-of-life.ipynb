Earth is the only world where life is confirmed. That fact can make astrobiology feel backward: we want to study life in the universe, but we only have one inhabited planet to examine in detail. The way forward is not to pretend that one example is enough to prove a universal rule. Instead, we use Earth life carefully. We ask what all known organisms share, what those shared features suggest about the basic requirements for living systems, and where those requirements might also exist beyond Earth.

The difficulty is that life is not one simple thing. A bacterium, a mushroom, a maple tree, a jellyfish, and a human are all alive, but they look and behave very differently. A car can move, use energy, turn, and release waste, but it is not alive. A wildfire can grow and spread, but it is not alive. A crystal can form an orderly structure, but it is not alive. This chapter builds a working definition of life by looking at several connected properties rather than searching for one magic word.

That working definition matters for the rest of the course. If we search only for intelligent civilizations, then most possible life is ignored. If we search only for animals and plants, then we also miss most of Earth history, when life was microbial. If we search for life in a scientific way, we need to understand cells, metabolism, heredity, evolution, and the environmental limits of organisms. Those ideas become the bridge between biology on Earth and the search for possible life elsewhere.

## Defining Life

### Why life is hard to define

A useful definition of life has to do two jobs at once. It has to include the things that are clearly alive, such as bacteria, plants, animals, fungi, and archaea. It also has to exclude things that merely imitate one or two properties of life. That second part is harder than it first appears.

A cat and a car are a good comparison. Both require energy to function. Both can move. Both can change direction. Both can release waste products. But the similarities break down once we ask where the organization comes from. A cat is made of cells, grows by using inherited instructions, repairs itself, responds internally to changing conditions, and comes from a lineage of reproducing organisms. A car is assembled externally. It may be complex, but it does not contain its own inherited program for making more cars.

Biologists therefore often describe life through a cluster of properties: order, reproduction, growth and development, energy utilization, response to the environment, and evolutionary adaptation. These properties are not independent boxes that can be checked one at a time. They work together. Order requires energy to maintain. Growth and reproduction require inherited information. Evolution requires reproduction with variation. A living organism is not just a collection of chemicals; it is a chemical system that keeps itself organized, uses energy, and belongs to a lineage that can change over generations.

### Order, reproduction, and development

All living organisms show some kind of order. The order is not just visible shape. It reaches down to the molecular level. In cells, molecules are arranged into membranes, proteins, chromosomes, ribosomes, and many other structures. Those structures make biological function possible. A cell membrane separates inside from outside. Proteins can act as structural supports, transporters, or catalysts. $\rm DNA$ stores information in a sequence of molecular building blocks. Without this kind of organization, the rest of life would not work.

Order is necessary, but it is not sufficient. A book has order. A quartz crystal has order. A snowflake has a striking pattern. None of these is alive. Even a purified $\rm DNA$ molecule or a single protein is not alive by itself, because it does not carry out the whole organized process of living. The lesson is that order matters only as part of a larger system.

Reproduction is also central to life, because evolution cannot operate unless information is passed from one generation to another. Reproduction can be sexual, involving two parents, or asexual, involving one parent. Many microbes reproduce by dividing into two daughter cells. Many plants and fungi can reproduce in multiple ways. The details vary, but reproduction connects individual organisms to a continuing lineage.

Even reproduction has complications. A mule is alive even though it is usually sterile. A worker ant may never reproduce individually, but it is part of a reproducing colony. Viruses can make copies of themselves only by using living cells. Prions, which are misfolded proteins, can cause other proteins to misfold in a similar way. These edge cases are why a one-sentence definition of life is not enough. Biology is full of systems that sit near the boundary.

Growth and development add another piece. Living organisms do not merely get larger. They develop patterns according to inherited instructions and environmental conditions. A seed becomes a plant by using stored information and available resources. A fertilized egg becomes an animal through a long sequence of regulated cell divisions and cell specializations. A wildfire also grows and spreads, but it does not develop according to hereditary information. Again, the property matters only when connected to the whole biological system.

### Boundary cases do useful work

The awkward cases near the boundary of life are not distractions. They help clarify what a definition is trying to do. Viruses are a good example. A virus can contain genetic material and can evolve over time, but it cannot reproduce by itself. It must enter a living cell and use that cell’s machinery. Outside a host cell, a virus is more like a package of biological instructions than an independently living organism.

```{figure-md} virus-boundary-case
<img src="https://upload.wikimedia.org/wikipedia/commons/5/52/Phage.jpg" alt="Electron micrograph of bacteriophages attached to a bacterial cell." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Viruses sit near the boundary of life because they contain biological information and evolve, but they require living cells to reproduce. Image source: Wikimedia Commons, Professor Graham Beards, *Phage.jpg*, CC BY-SA 3.0/GFDL.
```


Prions push the boundary from another direction. A prion is not a cell and does not contain $\rm DNA$ or $\rm RNA$. It is an incorrectly folded protein that can encourage other proteins to fold incorrectly. That is a kind of propagation, but it is not metabolism, cellular life, or heredity in the ordinary biological sense. Prions show why “makes more of itself” is not enough by itself.

Sterile organisms also show why reproduction has to be interpreted at the level of lineages, not isolated individuals. A mule is alive even if it usually cannot produce offspring. A human who never has children is still alive. Worker insects may not reproduce individually, but they can contribute to the reproduction of a colony. Life is not defined by whether one individual successfully reproduces at a given moment. It is defined by belonging to a biological system in which inherited information can be copied, varied, and sorted across generations.

These examples matter for astrobiology because a future discovery may not fit our first expectations. A chemical system on another world might show some life-like properties but not others. Scientists would then need to ask whether the system has organization, energy use, bounded chemistry, heredity, and evolution, rather than asking whether it looks familiar. A careful definition helps us avoid two opposite mistakes: rejecting unfamiliar life too quickly, or calling any interesting chemistry alive.

### Energy, disorder, and response

Living systems require energy because ordered systems do not maintain themselves for free. The second law of thermodynamics tells us that, when left alone, energy transformations tend to increase disorder. That does not mean life violates thermodynamics. It means life must constantly use energy from its surroundings to maintain local order.

```{figure-md} response-to-stimuli
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/b4d91acb1aec2baf52e9877fd4975016051161a8" alt="Sensitive plant with leaves that respond to touch." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Response to the environment is one property of life, but it becomes biological only when connected to metabolism, regulation, repair, heredity, and reproduction. Image source: OpenStax *Biology 2e*, [Themes and Concepts of Biology](https://openstax.org/books/biology-2e/pages/1-2-themes-and-concepts-of-biology), Figure 1.11 (credit: Alex Lomas).
```


A living organism can obtain energy from food, sunlight, or chemical reactions in its environment. That energy is used to build molecules, repair damage, move materials, reproduce, and maintain internal conditions. When the energy supply stops, the organized relationships inside the organism begin to fail. Cells lose the ability to repair damage. Chemical gradients break down. The organism eventually decays.

This idea is important for astrobiology because it separates the question of life from the question of sunlight. Sunlight is a powerful energy source for life at Earth’s surface, but it is not the only possible one. On Earth, some organisms use chemical reactions near deep-sea vents or hot springs. That means the search for life elsewhere should not be limited to sunny surfaces. Subsurface oceans, warm rock-water interfaces, or chemically active environments may also matter.

Living organisms also respond to their surroundings. Bacteria may move toward nutrients or away from harmful chemicals. Plants grow toward light. Mammals sweat, pant, shiver, or adjust blood flow to help regulate internal temperature. These responses are not merely passive. They are part of maintaining the conditions that allow life to continue.

Response by itself is still not enough. A thermostat responds to temperature changes, but it is not alive. A phone screen responds to touch, but it is not alive. The key difference is that biological response is integrated with metabolism, heredity, repair, growth, and reproduction. In a living cell, environmental response is part of the organism’s self-maintaining chemistry.

### Evolutionary adaptation and a practical definition of life

Evolutionary adaptation is the most important property for defining life because it explains why Earth life is so diverse. Evolution means change over time. In biology, the relevant change occurs across generations. A single organism can grow, learn, or respond during its lifetime, but biological evolution refers to changes in inherited traits in a population.

```{figure-md} darwin-finches
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/478dd2e5b1311ce62dca7fcd615c92c1c07fc4c3" alt="Galápagos finches with different beak shapes." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Darwin’s finches illustrate how inherited variation can be sorted by environment over generations. The important lesson is not the birds themselves, but the connection between variation, reproduction, and adaptation. Image source: OpenStax *Biology 2e*, [Understanding Evolution](https://openstax.org/books/biology-2e/pages/18-1-understanding-evolution), Figure 18.2.
```

Ideas about biological change long predate Darwin. Ancient Greek thinkers disagreed over whether living things could change through time, and early nineteenth-century naturalists were trying to explain fossils that did not look exactly like living species. Jean-Baptiste Lamarck argued that organisms adapt to their environments, which was an important step because it treated change through time as a natural process. The weakness was the mechanism. Lamarck is often summarized through inheritance of acquired characteristics: an organism changes during its lifetime and passes that acquired change to its offspring.



Darwin’s theory of evolution by natural selection made a different claim. Darwin emphasized overproduction and competition for survival: organisms often produce more offspring than can survive in a limited environment. He also emphasized individual variation: no two individuals are exactly alike, and some inherited differences affect survival or reproduction. Darwin’s conclusion was that unequal reproductive success can sort inherited variation. If a trait helps organisms leave more offspring in a particular environment, that trait can become more common over generations. Natural selection does not require organisms to plan ahead, and it does not mean individuals evolve during their lifetimes. Populations evolve as inherited traits become more or less common.

The contrast is important. Lamarck’s view puts the main change inside an individual during its lifetime. Darwin’s view puts the main change in a population across generations. Lamarck helped frame adaptation as a natural problem to explain, but Darwin supplied the mechanism that connected inherited variation, reproduction, and environmental sorting.

The Galápagos finches are a classic example because different populations became adapted to different foods and environments. Ground finches and tree finches did not become different because individual birds decided to change their beaks. Populations changed because inherited beak differences affected feeding success and reproduction under different island conditions. Artificial selection makes the same logic visible in a human-controlled setting. Selective breeding of plants and animals can strongly reshape traits over many generations because humans decide which organisms reproduce. Several familiar green vegetables, including cabbage, broccoli, kale, and Brussels sprouts, come from selective breeding of wild mustard.

Darwin did not know the molecular basis of heredity. Gregor Mendel’s pea-plant work later provided a way to think about inherited units, and modern genetics showed that $\rm DNA$ stores information that can be copied and modified. Evolution is a scientific theory because it has survived many tests, connects many lines of evidence, and has no credible alternative with comparable explanatory power. We can also watch evolution occur, as in viral variants such as alpha, delta, and omicron lineages of SARS-CoV-2. Those variants are not separate proof of all evolutionary history, but they are a direct example of inherited variation spreading through populations over time.

Biologists also use classification terms to keep related organisms organized. A species is a specific category of organisms that are genetically distinct from other groups; for sexually reproducing organisms, the term often also involves the ability to reproduce with one another. A genus is the broader category that contains one or more closely related species. Humans are classified as *Homo sapiens*: *Homo* is the genus and *sapiens* is the species.

A practical working definition is therefore: life is a system that can reproduce and evolve through natural selection. This definition is not perfect, but it is powerful. It emphasizes heredity, variation, and evolution rather than superficial appearance. It also helps astrobiology avoid looking only for organisms that resemble familiar plants and animals.

## Cells: The Basic Units of Life

### Cells as boundaries and chemical systems

A cell is a microscopic unit in which living matter is separated from the outside world by a membrane. That boundary is essential. Without a boundary, useful molecules would drift away, harmful materials could enter freely, and the chemical reactions needed for life would not be organized in one place.

```{figure-md} prokaryotic-cell
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/50163f8ff80f335574f41bfc10cc49a1e87cf9df" alt="Generalized prokaryotic cell showing membrane, cell wall, nucleoid region, ribosomes, and flagellum." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

A cell is not just a container. It is an organized boundary where genetic information, chemical reactions, and molecular machinery can stay concentrated enough to work together. Image source: OpenStax *Biology 2e*, [Prokaryotic Cells](https://openstax.org/books/biology-2e/pages/4-2-prokaryotic-cells), Figure 4.5.
```


Cells can live alone, as many bacteria and archaea do, or they can cooperate in multicellular organisms. In a multicellular organism, different cells specialize for different jobs. Muscle cells, nerve cells, leaf cells, and fungal cells do not look or behave the same way, but they all share the basic logic of cellular life: a controlled internal environment, inherited information, and metabolism.

$\rm DNA$ is housed inside cells. In many organisms, $\rm DNA$ sits inside a nucleus; in bacteria and archaea, it is not enclosed in the same kind of nucleus. Either way, heredity is not an abstract idea. It is stored in molecules inside cells and expressed through cellular chemistry.

Astrobiology has to be careful here. We should not assume that life elsewhere must use exactly the same cell structures or biochemistry as Earth life. But the cell teaches a general lesson that may be broader than Earth. Life probably needs some way to concentrate chemical reactions, protect useful molecules, and separate an internal system from the surrounding environment. A membrane-like boundary is one practical solution.

### Carbon chemistry and the possibility of alternatives

Life on Earth is built mostly from a small group of elements. Oxygen, carbon, hydrogen, and nitrogen make up most of the mass of typical cells. Oxygen is often bound with hydrogen in water. Carbon forms the backbone of organic molecules. Hydrogen and nitrogen are common in many biological molecules. Other elements, such as phosphorus, sulfur, calcium, sodium, potassium, magnesium, and iron, are also important even when they are present in smaller amounts.

Carbon is especially important because it can form four bonds. That lets carbon atoms build chains, rings, and branching structures. Carbon can also form double bonds, which expands the range of possible molecular shapes and reactions. This versatility helps explain why Earth life is carbon-based.

Science fiction often imagines silicon-based life, and the idea is not random. Silicon sits below carbon in the periodic table and can also form four bonds. But silicon has serious disadvantages in Earthlike chemistry. Comparable silicon bonds are generally weaker than carbon bonds. Silicon does not naturally form double bonds as easily as carbon. Silicon dioxide is a solid, while carbon dioxide is a gas, which makes waste removal and atmospheric cycling very different. These problems do not prove that silicon-based life is impossible under all conditions, but they explain why carbon is a more practical backbone for the only life we know.

### Common ancestry and why Earth life looks chemically unified

One of the most important facts about Earth life is that it is not a random collection of unrelated living systems. All known organisms share deep biochemical similarities. They use $\rm DNA$ and $\rm RNA$. They use related genetic codes. They use proteins built from amino acids. They use $\rm ATP$ as a major energy-transfer molecule. They use cells as the basic unit of life. These shared features are best explained by common ancestry.

```{figure-md} woese-tree-of-life
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/c1d242cd4ab1c5cea605d9074da958c8a9ee62d3" alt="Phylogenetic tree of life showing Bacteria, Archaea, and Eukarya." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Molecular comparisons reveal that Earth life is historically connected. The exact tree is a scientific model that can change with new data, but the deeper point is that organisms share ancestry rather than representing unrelated starts. Image source: OpenStax *Biology 2e*, [Themes and Concepts of Biology](https://openstax.org/books/biology-2e/pages/1-2-themes-and-concepts-of-biology), Figure 1.17 (credit: Eric Gaba; NASA Astrobiology Institute).
```

Common ancestry does not mean every organism is similar in size, shape, behavior, or environment. It means that living organisms inherited core molecular machinery from earlier life. Over billions of years, evolution produced enormous diversity, but it modified inherited systems rather than inventing every molecular process from scratch. This is why bacteria, oak trees, mushrooms, and humans can be so different while still sharing deep chemical patterns.

This point is important when thinking about extraterrestrial life. Earth’s biochemical unity is not proof that $\rm DNA$, $\rm ATP$, and proteins are universal requirements. They may be historical outcomes of life’s early evolution on Earth. A second origin of life elsewhere might use different molecules. At the same time, common ancestry teaches us that once a life system establishes a workable solution, later evolution tends to build on it. Biology is historical. It carries old solutions forward, sometimes long after those solutions first appeared. 

That historical character also affects evidence. When scientists compare $\rm DNA$ or $\rm RNA$ sequences among organisms, they are not only classifying life. They are reconstructing relationships. Molecular comparison turns heredity into a record of history, much as rocks and fossils do in geology. For this course, that connection matters: astrobiology often asks how present-day evidence can reveal a past process that we cannot watch directly.

### The molecular components of cells

Cells are built from complex organic molecules. Four groups are especially important: carbohydrates, lipids, proteins, and nucleic acids. These are not vocabulary words to memorize in isolation. They are functional categories that help explain what cells do.

Carbohydrates include sugars and starches. They can provide energy, and some carbohydrates help build structure. Cellulose, for example, helps support plant cell walls. Lipids include fats and membrane-forming molecules. They can store energy, but their role in membranes is just as important for astrobiology because membranes make cellular boundaries possible.

Proteins are working molecules. Some provide structure, while others act as enzymes. An enzyme is a catalyst, meaning it speeds up a chemical reaction without being used up by the reaction. Enzymes make life’s chemistry fast and specific enough to work under ordinary biological conditions. Proteins are built from amino acids. Many amino acids exist chemically, but Earth life uses a common set of about 20 most often. Another interesting feature is handedness: amino acids in living cells are overwhelmingly left-handed, while non-biological production usually creates mixtures of left- and right-handed forms.

### The major groupings of life on Earth

For much of history, people classified life by outward appearance. Plants and animals were the obvious categories because they were large enough to see. Microscopes changed that picture. In 1674, Antonie van Leeuwenhoek observed microscopic life, including bacteria-like organisms. Later classifications added groups such as Protista and Monera for microorganisms, and fungi were separated from plants because fungi feed and reproduce differently, often by spores. Microscopic life is also not just a curiosity. Most microbes are harmless to humans, and many are crucial to our survival. Bacteria in our intestines help break down and regulate nutrients as part of the microbiome; fermented foods such as yogurt are familiar everyday reminders that microbes can be useful.

The deeper revolution came from comparing organisms at the molecular level. Today, life is often organized into three domains: Bacteria, Archaea, and Eukarya. Bacteria and Archaea are entirely microbial. Eukarya includes many microbes as well as plants, animals, and fungi. Eukaryotic cells generally have nuclei, but appearance alone is not the main reason for separating the domains. Their membrane chemistry, protein-making machinery, and genetic details differ in important ways. Archaea can look like bacteria under a microscope, but in several molecular features Eukarya are more similar to Archaea than to Bacteria.

This matters for astrobiology because our intuition is biased toward large multicellular life. Trees, mushrooms, insects, fish, birds, and mammals are easy to notice. But on the tree of life, large multicellular organisms occupy only a few branches within Eukarya. Much of life’s diversity is microbial. If another world has life, microbes may be more likely than forests or animals, and they may also be harder to detect.

### Chirality and chemical clues

The handedness of biological molecules is a subtle but important clue. Some molecules come in left-handed and right-handed forms, much like your left and right hands have the same parts but cannot be perfectly placed on top of each other. Non-biological chemistry often produces mixtures of both handed forms. Earth life, however, strongly favors particular handedness in important molecules. Amino acids used in proteins are overwhelmingly left-handed, while sugars in nucleic acids have a consistent handedness of their own.

The broader lesson is that life can leave chemical patterns. A fossil bone is an obvious biological clue, but microscopic or chemical evidence may be more realistic for many astrobiology targets. If life ever existed on Mars, Europa, Enceladus, or another world, the evidence might be subtle: molecular distributions, isotope ratios, cell-like textures, or chemical disequilibria. Amino acids have been found in meteorites, and organic chemistry can occur in space and on planetary surfaces without biology. A scientific search must distinguish ingredients from activity, and activity from confirmed life. The strongest case would come from multiple independent lines of evidence pointing in the same direction.

## Metabolism: The Chemistry of Life

### What metabolism does

Metabolism is the blanket term for the chemical reactions that occur inside living cells. It includes reactions that break molecules down, reactions that build molecules up, and reactions that transfer energy from one form to another. A cell is not just a tiny bag of chemicals. It is more like a regulated chemical factory.

```{figure-md} energy-processing
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/bf69acd5fab5ca9a827c175f7d289b7be6d3dec8" alt="A condor in flight as an example of energy processing in an organism." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Metabolism connects energy to biological work. Even visible activities such as movement depend on chemical reactions inside cells that capture, transfer, and spend energy. Image source: OpenStax *Biology 2e*, [Themes and Concepts of Biology](https://openstax.org/books/biology-2e/pages/1-2-themes-and-concepts-of-biology), Figure 1.14 (credit: Pacific Southwest Region U.S. Fish and Wildlife Service).
```

The factory analogy helps if we use it carefully. A factory needs raw materials, energy, instructions, and machinery. A cell needs simple molecules that can be rearranged into more complex organic molecules. It needs energy to drive reactions that would not happen easily on their own. It needs inherited information to specify the proteins and $\rm RNA$ molecules that perform cellular jobs. It also needs catalysts, especially enzymes, to make reactions occur at useful rates.

Cells are versatile because they can build a wide variety of products from a limited set of starting materials. That versatility is a product of evolution. Enzymes did not appear because a cell understood chemistry and designed a solution. They are inherited molecular tools shaped by natural selection over many generations.

For the search for life, metabolism gives us a practical way to think about habitability. A possible habitat must provide usable raw materials and usable energy. It also likely needs a medium, such as liquid water on Earth, where molecules can dissolve, move, meet, and react.

### ATP and the energy cycle

Earth life uses adenosine triphosphate, or $\rm ATP$, as a common energy-transfer molecule. $\rm ATP$ is not the ultimate source of all energy. Instead, it acts like a short-term energy currency inside cells. Energy from food, sunlight, or chemical reactions can be used to convert $\rm ADP$ plus a phosphate group into $\rm ATP$. Later, $\rm ATP$ can break back down into $\rm ADP$ plus phosphate, releasing energy that can drive cellular work.

```{figure-md} atp-structure
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/aa87a8b51a1ad240a078f55cb1038b18a560aabd" alt="Molecular structure of adenosine triphosphate showing adenine, ribose, and three phosphate groups." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

$\rm ATP$ is often described as an energy-transfer molecule because cells can use reactions involving its phosphate groups to couple energy release to cellular work. Image source: OpenStax *Biology 2e*, [ATP: Adenosine Triphosphate](https://openstax.org/books/biology-2e/pages/6-4-atp-adenosine-triphosphate), Figure 6.13.
```

The simplified cycle is

$${\rm ADP} + {\rm P} + {\rm energy} \rightarrow {\rm ATP},$$

for storage, and

$${\rm ATP} \rightarrow {\rm ADP} + {\rm P} + {\rm usable\ energy},$$

for release. The real cellular chemistry is more detailed, but this simplified picture captures the central point: $\rm ATP$ helps connect energy-producing reactions to energy-requiring reactions.

Students often think of microbes as “simple” life and animals as “advanced” life. That framing is misleading. Microbes are small, but they are not failed attempts at becoming animals. They are highly successful organisms adapted to their own ways of life. Bacteria and archaea have occupied Earth for billions of years, and they still dominate many environments by abundance, diversity, and metabolic variety.

### Carbon sources, energy sources, and metabolic categories

Organisms need carbon because carbon forms the backbone of organic molecules. They also need energy because building and maintaining those molecules requires work. The source of carbon and the source of energy are not always the same.

The word part *troph* means feeding or nourishment. An autotroph obtains carbon from the environment, usually from carbon dioxide or related inorganic sources. A heterotroph obtains carbon from organic compounds, which usually means consuming other organisms or their products. The prefixes describe energy source. A phototroph uses light. A chemotroph uses chemical reactions.

Putting those terms together creates four broad categories. Photoautotrophs, such as plants, algae, and cyanobacteria, obtain carbon from the environment and use sunlight as an energy source. Chemoautotrophs obtain carbon from inorganic sources and use energy from chemical reactions, such as reactions involving sulfur, iron, hydrogen, or ammonia. *Sulfolobus*, an archaeon found in sulfur-rich hot springs, is a useful example of a chemoautotroph. Photoheterotrophs use light for energy but obtain carbon from organic compounds; *Chloroflexus* is an example that can use photosynthesis for energy while obtaining carbon from other bacteria in salty or hot environments. Chemoheterotrophs obtain both carbon and energy from organic compounds; humans, other animals, fungi, and many microbes are familiar examples.

## DNA and Heredity

### DNA structure and replication

Life needs structure and metabolism, but it also needs heredity. Without heredity, there is no reliable way to pass useful information from one generation to the next. Without variation in heredity, evolution has nothing to sort. $\rm DNA$ is the main hereditary molecule for Earth life.

$\rm DNA$ has a double-helix structure, often compared to a twisted ladder or zipper. The outside edges are sugar-phosphate backbones. The rungs are pairs of bases. There are four $\rm DNA$ bases: adenine, thymine, cytosine, and guanine, usually abbreviated A, T, C, and G. The pairing rules are specific: A pairs with T, and C pairs with G.

Those pairing rules are what make $\rm DNA$ copyable. When $\rm DNA$ replication begins, the two strands separate. Each original strand can then act as a template for a new complementary strand. If one side has an A, the new partner must be T. If one side has a C, the new partner must be G. This is the basic logic behind high-fidelity copying.

The phrase “basic logic” hides enormous molecular complexity. In real cells, $\rm DNA$ replication requires more than a dozen specialized enzymes and helper proteins that unzip the molecule, stabilize the separated strands, copy new strands onto the exposed templates, check for errors, and help rezip or repair the final product. Complementary pairing makes high-fidelity copying possible, but replication is not perfect. Errors sometimes occur, and those errors are one source of mutation. The complexity of the modern system is also one reason many origin-of-life scenarios treat $\rm RNA$ or $\rm RNA$-like molecules as possible predecessors to modern $\rm DNA$ heredity.

### DNA as encoded information

The operating instructions in $\rm DNA$ come from the order of the bases. A short sequence such as ACTCAGCTTCAAGGG is not meaningful because of the letters themselves. It is meaningful because cellular machinery reads the sequence according to a genetic code. The full sequence of bases in an organism is its genome. Individual instructions or functional units are called genes, although not all important $\rm DNA$ is part of a protein-coding gene.

The human genome also shows why “more $\rm DNA$” does not simply mean “more advanced.” The first human genome sequence took about 13 years and roughly 3 billion dollars to complete, ending in 2003. Modern sequencing can cost hundreds of dollars and can be completed far more quickly. The human genome has only about 25,000 genes, while rice has about 37,000. The number of genes alone does not tell us how complex an organism is, because gene expression, regulation, timing, and cellular context matter enormously. Some noncoding $\rm DNA$ was once casually called “junk” $\rm DNA$ because it did not code directly for proteins, but some of it has regulatory roles, some records evolutionary history, and some remains poorly understood. 

```{figure-md} genetic-code-table
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/d99b2d20bb12abe0a633a49c567152957e014db7" alt="Genetic code table showing codons and the amino acids or stop signals they specify." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

The genetic code turns nucleotide sequences into protein-building instructions. Its redundancy helps explain why some mutations change a sequence without changing the final amino acid. Image source: OpenStax *Biology 2e*, [The Genetic Code](https://openstax.org/books/biology-2e/pages/15-1-the-genetic-code), Figure 15.4 (credit: modification of work by NIH).
```

The code is read in three-base words called codons. With four possible bases and three positions per word, there are $4^3 = 64$ possible codons. This is more than the 20 amino acids commonly used in proteins, so the code contains redundancy. For example, several codons can specify the same amino acid. That redundancy matters because some mutations can change a $\rm DNA$ word without changing the amino acid that is placed into the protein.

A two-base code would produce only $4^2 = 16$ possible words, which is not enough to specify 20 amino acids without a more complicated scheme. A three-base code gives more room. This simple counting argument helps students see why the genetic code has the structure it does.

```{admonition} Try the numbers
:class: note

The number of possible $\rm DNA$ words depends on how many base choices exist and how many positions are in each word. With $4$ bases and $n$ positions, the number of possible words is

$$N = 4^n.$$

For two-base words, $N = 4^2 = 16$. For three-base words, $N = 4^3 = 64$. This is why a three-base genetic code has enough possible words to encode the usual 20 amino acids with redundancy.
```

`````python
# Count possible DNA code words for two-base and three-base words.
number_of_bases = 4       # A, T, C, and G are the four DNA bases
two_base_word_length = 2  # number of positions in a two-base word
three_base_word_length = 3 # number of positions in a three-base word

words_with_two_bases = number_of_bases ** two_base_word_length
words_with_three_bases = number_of_bases ** three_base_word_length

print(f"A two-base genetic word can make {words_with_two_bases} possible words.")
print(f"A three-base genetic word can make {words_with_three_bases} possible words.")
print("A three-base code has enough possible words to encode 20 common amino acids with redundancy.")
`````

$\rm RNA$ carries out many of the instructions stored in $\rm DNA$. $\rm RNA$ is similar to a single strand of $\rm DNA$, but it uses uracil, abbreviated U, where $\rm DNA$ uses thymine, so its base alphabet is A, C, G, and U. Messenger $\rm RNA$, or $\rm mRNA$, carries a copied message from $\rm DNA$ to the ribosome. Transfer $\rm RNA$, or $\rm tRNA$, brings amino acids to the ribosome. Ribosomal $\rm RNA$, or $\rm rRNA$, is part of the ribosome itself and helps build proteins. The overall process of turning nucleic-acid information into a protein is called translation.

This system is another sign of common ancestry. The genetic code is nearly universal among Earth organisms, with only minor variations. That does not mean the code was inevitable. It means that once an early coding system became established in the common ancestry of Earth life, later organisms inherited and modified it rather than starting over from scratch.

### From genes to proteins without treating DNA as magic

It is tempting to say “$\rm DNA$ is the blueprint for life,” but that phrase can hide the process. $\rm DNA$ does not directly build a cell the way a blueprint directly guides a construction crew. $\rm DNA$ stores sequences. Cellular machinery reads those sequences. $\rm RNA$ molecules carry and help interpret the instructions. Ribosomes assemble amino acids into proteins. Proteins then do much of the structural and chemical work of the cell.

```{figure-md} central-dogma
<img src="https://openstax.org/apps/image-cdn/v1/f%3Dwebp/apps/archive/20260407.195030/resources/abdd78d09479f6d641ea2add24d0634f8bd27358" alt="Diagram of genetic information flowing from DNA to RNA to protein." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

The phrase “$\rm DNA$ is a blueprint” is only a shortcut. Cells must copy, process, and translate molecular information through $\rm RNA$, ribosomes, and proteins for heredity to affect structure and function. Image source: OpenStax *Biology 2e*, [The Genetic Code](https://openstax.org/books/biology-2e/pages/15-1-the-genetic-code), Figure 15.3.
```

A better analogy is an operating system plus a manufacturing system. $\rm DNA$ stores instructions, but the instructions only matter because the cell has machinery that can interpret them. This is one reason origin-of-life research is difficult. A modern cell depends on a tightly connected system: $\rm DNA$ stores information, $\rm RNA$ helps express that information, proteins perform many of the jobs, and enzymes are needed to copy and maintain the information. Explaining how such a system began requires understanding possible simpler predecessors.

$\rm RNA$ is important in this discussion because it can both store information and participate in chemical activity. Many origin-of-life scenarios therefore consider an earlier stage in which $\rm RNA$-like molecules played a larger role before the modern $\rm DNA$-$\rm RNA$-protein system became established. This is not presented as a finished answer to the origin of life. It is an example of how scientists try to build plausible intermediate steps rather than jumping from simple chemistry to modern cells in one leap.

This also helps explain why heredity is tied to evolution. A molecule that stores information but is never copied cannot support biological evolution. A molecule that is copied perfectly forever creates continuity but no variation. A molecule that is copied with some fidelity and occasional variation creates the possibility of descent with modification. Life needs enough accuracy to preserve useful information and enough imperfection to allow populations to change.

### Mutation and evolution

$\rm DNA$ copying is accurate, but not perfect. A useful round number is that copying errors occur at roughly one error per billion bases copied. Errors can take several forms. The wrong base can be inserted. An extra base can be added. A base can be deleted. A whole sequence can be duplicated or removed. High-energy radiation, such as ultraviolet light, and some chemicals can also damage $\rm DNA$.

Any change in $\rm DNA$ is a mutation. A mutation is not automatically harmful. Some mutations are neutral because they do not change the amino acid sequence or because the change does not strongly affect the organism. Some are harmful because they disrupt a critical protein or regulatory process. A few can be beneficial in a particular environment.

The effect of a mutation depends on context. Some changes have little or no effect. For example, a codon change from $\rm ACC$ to $\rm ACA$ can still specify the same amino acid, threonine. Other changes alter a protein more directly. Sickle-cell anemia can be traced to a single base change in the relevant gene, often summarized as an A-to-T substitution in the $\rm DNA$ sequence. Many mutations are harmful or lethal because they prevent the production of a critical protein or disrupt when and where a gene is used. A change that is harmful in one environment might be less harmful, neutral, or even helpful in another. Sickle-cell-related genetic variation is a classic example of context dependence because it can cause disease in some genetic combinations but also offers some protection against malaria in certain environments. The general lesson is that evolution depends on variation, and mutation is one major source of that variation.

Mutation also helps explain why the genetic code’s lack of spaces or punctuation matters. In English, “the fat cat ate the rat” is readable because spaces mark the words. If the spaces are removed, *thefatcatatetherat* can still be decoded only if the reader knows where to group the letters. If a letter is inserted near the beginning, the grouping can shift: *the***a***fatcatatetherat* becomes difficult to parse in the original three-letter groups. Similarly, insertions or deletions in $\rm DNA$ can shift how codons are read, potentially changing many amino acids downstream.

```{admonition} Try the numbers
:class: note

Suppose a simplified genome has $3.2\times 10^9$ bases and the copying error rate is about $1$ error per $10^9$ bases. The expected number of errors per full copy is

\begin{align*}
N_{\rm errors} &= (3.2\times 10^9\ {\rm bases})\left(\frac{1\ {\rm error}}{10^9\ {\rm bases}}\right) \\
&= 3.2\ {\rm errors}.
\end{align*}

That is a small rate per base, but a large genome contains many bases. High accuracy still leaves room for occasional variation.
```

`````python
# Estimate the expected number of copying errors in a simplified genome.
genome_size_bases = 3.2e9       # number of bases in the simplified genome
error_rate_per_base = 1 / 1e9   # copying errors per base copied

expected_errors = genome_size_bases * error_rate_per_base

print(f"For a simplified genome with {genome_size_bases:.1e} bases, the expected number of copying errors is {expected_errors:.1f}.")
print("This small per-base error rate can still create a few differences each time a very large genome is copied.")
`````

If a mutation occurs in a cell lineage that contributes to offspring, it can be inherited. Once inherited variation exists, natural selection can act. Mutations provide the raw material for evolution, but they do not by themselves guarantee adaptation. Many mutations disappear. Some persist by chance. Some spread because they help organisms reproduce in a particular environment. A human trait such as red hair can become common in a population when a mutation is inherited and then amplified by population history, isolation, and reproduction.

Most inheritance is vertical gene transfer, from parent to offspring. Microbes also complicate that simple parent-to-offspring picture. Bacteria can sometimes transfer genetic material sideways between organisms, a process called horizontal or lateral gene transfer. This is one way antibiotic resistance can spread. It is also one reason the microbial tree of life is more complicated than a simple branching family tree. Humans also use this capacity in biotechnology: bacteria can be given genetic instructions to produce human insulin, which can then be purified for medicine.

For astrobiology, heredity is central because it gives us a way to think about life as a historical process. A living system is not just chemistry happening now. It is chemistry connected to past generations and capable of changing in future generations.

## Life at the Extreme

### Extremophiles and the range of habitability

An extremophile is an organism that lives outside conditions humans would consider ordinary. The word combines “extreme” with *phile*, meaning lover. Thermophiles live in hot environments. Hyperthermophiles live at even higher temperatures. Other organisms tolerate extreme dryness, acidity, alkalinity, salinity, cold, pressure, or radiation.

Deep-sea hydrothermal vents are a powerful example. Water can reach temperatures above the normal boiling point, but high pressure keeps it liquid. Some microbes living near vents are chemoautotrophs: they obtain carbon from dissolved carbon dioxide and energy from inorganic chemical reactions in the hot water. *Pyrolobus fumarii*, whose name means “fire lobe of the chimney,” is an example of a hyperthermophile associated with hot vent environments. This matters because it shows that sunlight is not the only path to a biosphere.

Hot springs are another useful example because they are easier to observe directly. Thermophiles in hot springs have enzymes adapted to high temperatures. Those enzymes may fail at ordinary temperatures, just as our enzymes fail at extreme heat. Some extremophiles are anaerobic, meaning they live without oxygen and may be poisoned by oxygen-rich environments. From their point of view, our comfortable atmosphere could be hostile.

```{admonition} Checkpoint
:class: tip margin checkpoint

Why is an environment that seems extreme to humans not automatically impossible for life?
```

Other examples broaden the lesson. Microbes can survive in the Atacama Desert, one of the driest places on Earth, where organisms face dryness, temperature swings, and high ultraviolet radiation. Some survive inside rocks in the cold dry valleys of Antarctica. Some tolerate high acidity, alkalinity, or salt. *Deinococcus radiodurans* can survive radiation levels far beyond what would kill humans and most animals, and it has even survived exposure experiments relevant to space conditions. Endospores can act as resting cells that neither grow nor die under harsh conditions. *Bacillus anthracis* can survive as durable spores through lack of water, extreme heat or cold, and many chemical stresses. Tardigrades, or “water bears,” can survive many extreme conditions, although they are not true extremophiles in the same sense because they generally do better in more ordinary environments.

### Are extremophiles really extreme?

The word extreme depends on the observer. From our point of view, boiling water, acidic pools, deep rock, and intense radiation seem extreme. But an organism adapted to one of those settings is not living at the edge of its own normal. It is living in the conditions its biology can handle.

```{figure-md} grand-prismatic-spring
<img src="https://astrobiology.nasa.gov/uploads/filer_public_thumbnails/filer_public/bb/1e/bb1e0ef9-23e5-4aca-809a-b19585cfd9df/hero_grand_prismatic_spring.jpg__1240x510_q85_crop_subject_location-620%2C254_subsampling-2.jpg" alt="Grand Prismatic Spring at Yellowstone with colorful microbial mats." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Image source: NASA Astrobiology; photo credit: Jim Peaco/National Park Service.
```

This perspective is important because Earth’s history has not always looked like the present. Oxygen-rich air, mild surface temperatures, and large animals are relatively recent features in the full history of life. For much of Earth history, life was microbial. Many early environments may have been hot, chemically harsh, low in oxygen, or otherwise unlike the settings humans prefer. 

Extremophiles therefore matter for the search for life in the Solar System and beyond, including SETI in its broad sense as a search for life or life-related signals beyond Earth. They suggest that possible habitats may include subsurface oceans, ice-covered environments, salty brines, hydrothermal systems, and rock-water interfaces. Because some extremophile lifestyles may resemble conditions common on early Earth, they also remind us that the first widespread Earth life was probably microbial rather than animal-like. They help us think about what evidence of life might look like when organisms are microscopic and hidden.

At the same time, extremophiles should not make us careless. The existence of organisms in harsh Earth environments does not mean every harsh environment is habitable. Life still needs usable chemistry, available energy, a suitable medium, and enough stability over time. The astrobiological lesson is not “life can live anywhere.” It is more careful and more useful: life can live in more kinds of environments than human intuition suggests.

### Extremophiles and the danger of overcorrecting

Extremophiles are exciting because they expand our imagination, but they can also tempt us into a different mistake. After learning about organisms that survive boiling water, intense radiation, deep rock, or extreme dryness, it is easy to say that life can survive almost anything. That statement goes too far.

```{figure-md} atacama-rover
<img src="https://astrobiology.nasa.gov/uploads/filer_public/d6/bf/d6bfd332-99ea-40c4-ba66-2b16dd4fca68/193578_web.jpg" alt="A solar-powered rover tested in the Mars-like Atacama Desert." style="max-height:50vh; width:auto; max-width:100%; display:block; margin-left:auto; margin-right:auto;">

Earth analog sites help scientists test instruments and search strategies, but an analog is not a duplicate of another world. The Atacama Desert is useful because it is dry and Mars-like in some ways, not because it proves Mars is inhabited. Image source: NASA Astrobiology; photo credit: Prof. Stephen B. Pointing.
```

Survival is not the same as growth. Some organisms can survive harsh conditions in a dormant state, but they may not actively grow or reproduce there. Tardigrades are a good example. They can survive remarkable stresses, especially in dormant forms, but that does not mean they prefer those environments or can complete their life cycle there. Endospores are also survival structures, not active thriving cells.

For astrobiology, this distinction matters. A spacecraft might find an organism-like particle that survived transport, but survival alone would not prove that the environment where it was found supports an active biosphere. A true habitat must allow metabolism, repair, and reproduction over time. The question is not only “Can life endure this?” but “Can life live, grow, and evolve here?”

This is why Earth analogs must be used carefully. The Atacama Desert, Antarctic dry valleys, hydrothermal vents, acidic springs, and salty lakes can help scientists test instruments and hypotheses. They show what kinds of evidence life may leave under difficult conditions. But no Earth analog is a perfect copy of Mars, Europa, Enceladus, Titan, or an exoplanet. Analog environments are tools for thinking, not proof that another world is inhabited.

The careful middle position is the strongest one. Extremophiles show that human comfort is a poor guide to habitability. They also show that life still has limits. A global catastrophe that would be devastating for humans and many ecosystems would not necessarily erase every microbial lineage; life is more robust than our everyday experience suggests. But that robustness is not the same as invincibility, and it does not make every environment inhabited. The scientific search for life beyond Earth lives in that tension: keep the search broad enough to avoid Earth-surface bias, but constrained enough that evidence remains meaningful.
