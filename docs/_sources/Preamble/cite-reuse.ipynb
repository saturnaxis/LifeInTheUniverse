## Suggested Citation

If you use or adapt material from this book, please cite it as an online book. A recommended citation format is:

> Quarles, B. (Year). *Life in the Universe*. East Texas A&M University.  
> Jupyter Book. URL

For example:

> Quarles, B. (2026). *Life in the Universe*. East Texas A&M University.  
> Jupyter Book. https://github.com/saturnaxis/ModernAstro

If you cite a specific chapter, section, figure, table, or exercise, include the section title and provide the direct URL when possible.

Because this book is a living document, you may also include the access date.
