For example, suppose we want to estimate how long sunlight takes to reach Saturn. We can start with a known value: sunlight takes about $8.2\ {\rm min}$ to travel from the Sun to Earth. Earth’s average distance from the Sun is $1\ {\rm AU}$, while Saturn’s average distance from the Sun is about $9.5\ {\rm AU}$.

Because light travels at a constant speed, the travel time scales directly with distance. Saturn is about $9.5$ times farther from the Sun than Earth, so the light travel time is also about $9.5$ times longer.

\begin{align*}
t_{\rm Saturn} &= \left(\frac{a_{\rm Saturn}}{a_{\oplus}}\right)t_{\oplus} \\
&= \left(\frac{9.5\ {\rm AU}}{1\ {\rm AU}}\right)(8.2\ {\rm min}) \\
&= 77.9\ {\rm min}.
\end{align*}

This means sunlight takes about $78\ {\rm min}$, or about $1.3\ {\rm hr}$, to reach Saturn. The Python version repeats the same scaling calculation.

````python
# Light travel time from the Sun to Earth in minutes
earth_light_time = 8.2

# Average orbital distances in AU
earth_distance = 1.0
saturn_distance = 9.5

# Scale the Earth light travel time by the distance ratio
saturn_light_time = (saturn_distance / earth_distance) * earth_light_time

# Convert the light travel time from minutes to hours
saturn_light_time_hours = saturn_light_time / 60

print(f"Sunlight takes about {saturn_light_time:.1f} minutes to reach Saturn.")
print(f"This is about {saturn_light_time_hours:.1f} hours.")
````

The important idea is the scaling. Once we know the light travel time for $1\ {\rm AU}$, we can estimate the light travel time to another planet by multiplying by that planet’s distance in AU.
