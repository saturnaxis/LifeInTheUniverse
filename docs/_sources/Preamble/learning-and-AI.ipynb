# Learning with AI and NotebookLM

AI tools can be useful for learning astronomy when they are used carefully. They can summarize a difficult section, generate practice questions, explain vocabulary, help compare ideas, or give feedback on your understanding.

However, AI tools should not replace your own thinking. They can make mistakes, oversimplify an idea, or give an answer that sounds confident but is not well supported. In this course, AI should be used as a study partner, not as an authority.

NotebookLM is especially useful because it can work directly with the course notes you provide as sources. You are encouraged to use NotebookLM with these notes to review sections, generate study guides, create practice questions, and ask for explanations of difficult concepts.

```{admonition} A good AI prompt has three parts
:class: note

A strong learning prompt usually includes:

1. **Persona** — tell the AI what role to take.
2. **Context** — tell the AI what you are studying and what you already know.
3. **Task** — tell the AI exactly what you want it to do.

This structure helps the AI give a response that is more useful for learning.
```

For example, instead of writing:

> Explain habitable zones.

write:

> Act as a patient introductory astronomy tutor. I am an ASTR120 student learning about habitable zones for the first time. Explain what a habitable zone is, why it does not guarantee that life exists, and give me three checkpoint questions to test my understanding.

The second prompt is better because it gives the AI a role, a context, and a specific task.

---

## Example AI Prompts for Learning

### Summarizing a section

> Act as an introductory astronomy tutor. I am reading a section about the search for life beyond Earth. Summarize the section in plain language, keep the important science vocabulary, and list the three most important ideas I should remember.

### Checking understanding

> Act as a study coach for an ASTR120 student. I just read a section about evidence and scientific claims. Ask me five conceptual questions, one at a time. After each answer, give feedback and help me improve my explanation.

### Explaining vocabulary

> Act as a patient science tutor. I am confused by the terms hypothesis, evidence, model, and theory. Explain each term using simple examples from astronomy or astrobiology.

### Preparing for a quiz

> Act as a quiz generator for an introductory astronomy course. Based on this section of notes, create ten multiple-choice questions. Do not show the answers until after the questions. Then explain why the correct answers are correct.

### Reviewing a confusing idea

> Act as an astronomy instructor. I am confused about why detecting oxygen in an exoplanet atmosphere would be interesting but would not automatically prove life exists. Explain the idea carefully and include at least two possible nonliving explanations.

### Improving a written answer

> Act as a helpful writing coach for an introductory astronomy student. Read my paragraph and give feedback on scientific accuracy, clarity, and use of evidence. Do not rewrite the paragraph for me. Instead, give suggestions that help me revise it myself.

---

## Using NotebookLM with These Notes

NotebookLM works best when you give it the course notes as a source. After adding the relevant chapter or section, you can ask it to help you study from that material.

Good NotebookLM prompts include:

- Summarize this section in plain language, but keep the important astronomy vocabulary.
- Create a study guide for this section.
- List the main claims in this section and describe the evidence for each one.
- Generate five checkpoint questions that test whether I understood the section.
- Explain this section as if I am new to astronomy.
- What are common misconceptions about this topic?
- Turn this section into a short review quiz.
- Help me compare two ideas from this section.
- Identify the most important figures, examples, or calculations in this section.
- Give me a list of questions I should be able to answer after reading this section.

```{admonition} Use AI responsibly
:class: warning

AI can help you study, but it can also be wrong. Always compare AI-generated explanations with the course notes, class discussion, and reliable scientific sources.

Do not copy AI-generated text as your own work. Use AI to improve your understanding, not to avoid thinking.
```
