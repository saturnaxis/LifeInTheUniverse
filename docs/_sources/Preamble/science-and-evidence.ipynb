# Science and Evidence

Science is a way of learning about the natural world by testing ideas against evidence. A scientific claim should not be accepted simply because it sounds reasonable, comes from an authority, or agrees with what we already believe. Every scientific claim requires evidence.

However, evidence does not usually prove that a claim is absolutely true. Instead, evidence changes how confident we should be. Strong evidence can make a claim more likely. Weak or conflicting evidence can make a claim less likely. New evidence can also force scientists to revise or reject an earlier explanation.

This is one reason science is powerful. Scientific knowledge is not a list of final answers. It is a process for building explanations that are supported by observation, measurement, calculation, and repeated testing. The best scientific explanations are the ones that account for the evidence we have while continuing to survive new tests.

In this book, we will often connect large questions to specific evidence. How do we know the age of the Solar System? How do we know what stars and planets are made of? How can we infer conditions on worlds we have never visited? These questions require more than imagination. They require observations, models, measurements, and careful reasoning.

```{admonition} Evidence and confidence
:class: note

A useful way to think about science is:

1. A claim begins as an explanation or hypothesis.
2. Evidence is gathered through observation, measurement, experiment, or calculation.
3. Each piece of evidence increases, decreases, or complicates our confidence in the claim.
4. Strong claims require strong, repeated, and independent evidence.

Science rarely asks, “Can this be proven with perfect certainty?” Instead, it asks, “How well does the evidence support this claim compared with other possible explanations?”
```

Scientific reasoning also requires humility. A single observation may have more than one explanation. A measurement may have uncertainty. A model may leave out important details. For this reason, scientists do not only ask whether evidence supports a claim. They also ask what else could explain the same evidence.

```{admonition} How to evaluate a scientific claim
:class: tip

When you encounter a scientific claim, ask yourself:

1. What evidence supports the claim?
2. How was that evidence measured or observed?
3. Are there other explanations for the same evidence?
4. What additional evidence would make the claim stronger or weaker?

These questions help separate scientific reasoning from guessing, speculation, or opinion.
```
